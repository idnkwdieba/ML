Построим гистограммы для числовых признаков:

```python
numeric_colums = data.select_dtypes(include=['float', 'int']).columns.tolist()
data_hist = data[numeric_colums]
fig, axes = plt.subplots(nrows=6, ncols=3, figsize=(12, 12))
axes = axes.ravel()
for i in range(len(data_hist.columns)):
    axes[i].hist(data_hist.iloc[:, i], bins=10)
    axes[i].set_title(data_hist.columns[i])
plt.tight_layout()
plt.show()
```

Построим матрицу корреляции:

```python
plt.figure(figsize=(16, 16))
sea.heatmap(round(data_hist.corr(), 1), annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.show()
```

По матрице корреляции можем видеть, что: 

люди с ожирением имеют большой вес (логично);

более зрелые люди чаще имеют собственное транспортное средство;

люди со случаями ожирения в семье чаще страдают ожирением;

мужчины выше женщин;

высокие люди имеют больший вес;

более зрелые люди меньше пользуются гаджетами;

люди с ожирением не обязательно делают перекусы между основными приемами пищи.
